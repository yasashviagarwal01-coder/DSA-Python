```python
#IMP STEP
movie_id=None
movie_list=[]

for customer in df['Cust_ID']:
  if ":" in customer:
    movie_id=int(customer.replace(":",""))

  movie_list.append(movie_id)
```

```python
#Now we add in the dataframe
df['Movie_Id']=movie_list
df
```

```python
#Now here'
#Everything is good the rows with null values are not required.
df.dropna(inplace=True)
```

```python
#Lets convert obj into int
df['Cust_ID']=df['Cust_ID'].astype('int')
```

```python
#So if give this much data it will take lot of time'
#So we will filter the data ---> But how?
#less no.of ratings ---> drop
#Customers who rateed less movies ---> drop

#But how can we do this?
#percentile
```

```python
#so for this i need to get count of rtaings for each movie

movie_rat_count=df['Movie_Id'].value_counts()
movie_rat_count
```

```python
bench_marks1=round(movie_rat_count.quantile(0.6),0)
bench_marks1


#so now if a movie ratings count having below 908 i gonna drop them
```

```python
drop_movie_index=movie_rat_count[movie_rat_count<bench_marks1].index
drop_movie_index

# so here i got the movie ids that i want to drop
```

```python
#2nd filter
cust_rat_count=df['Cust_ID'].value_counts()
cust_rat_count
```

```python
bench_marks2=round(cust_rat_count.quantile(0.6),0)
bench_marks2
```

```python
drop_cust_index=cust_rat_count[cust_rat_count<bench_marks2].index
drop_cust_index
```

```python
drop_movie_index,drop_cust_index
```

```python
df=df[~df['Movie_Id'].isin(drop_movie_index)]
df=df[~df['Cust_ID'].isin(drop_cust_index)]
```

```python
from surprise import Reader,Dataset,SVD
from surprise.model_selection import cross_validate


#Reader ---tells surprise how to read your dataset
#dataset ----> puts your data into surprise's box
          # ---->It converts CSV/DF into something surprise understands.
#SVD - tool
```

```python
reader = Reader()
```

```python
data=Dataset.load_from_df(df[['Movie_Id','Cust_ID','Ratings']][:100000],reader)
#We are converting a pandas DF into surprise's internal format so SVd can learn from it.
```

```python
data
```

```python
model=SVD()
```

```python
cross_validate(model,data,measures=['RMSE'],cv=3)
```
